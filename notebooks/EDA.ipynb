```python
import pandas as pd
import plotly.express as px

fastfood_data = pd.read_csv("/Users/andrearmz/Documents/Triple Ten/Sprint 7/Sprint-7-project/fastfood_data.csv") # leer los datos

# Paso 3. Análisis exploratorio de datos

histogram = px.histogram(fastfood_data, x="Company") # crear un histograma
histogram.show()

scatter = px.scatter(fastfood_data, x="Item", y="Calories") # crear un gráfico de dispersión
scatter.show()

pie_chart = px.pie(fastfood_data, values='Calories', names='Item')
pie_chart.show()

fastfood_data.info() # obtener información de los datos
```
